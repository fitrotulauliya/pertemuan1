```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
x = np.arange(0,2,0.05)*np.pi
y = np.sin(x)
plt.plot(x,y)

y2 = np.cos(x)
plt.plot(x,y2,'r')

y3 = np.sin(2*x)
plt.plot(x,y3,'go-')

y4 = np.cos(2*x)
plt.plot(x,y4,marker='x',color='orange')

plt.show()
```

```python
x = np.arange(0,2,0.05)*np.pi
y = np.sin(x)
plt.plot (x,y, label=r'$ \mathcal{y = sin(x)} $')

y2 = np.cos(x)
plt.plot (x,y2,'r', label=r'$ \mathcal{y = cos(x)} $')

plt.legend(loc='upper center')
plt.show()
```

```python
#cara menambahkan teks dalam graik adalah dengan cara memanggil fungsi berikut :
import matplotlib.pyplot as plt
%matplotlib inline

nama_desa_di_wilayah_karangploso = ['kepuharjo','ngijo','kendalsari','supiturang','kasin',]
jumlah_penduduk = ['87','88','95','77','99']

plt.figure(figsize=(7,3))
plt.bar(nama_desa_di_wilayah_karangploso, jumlah_penduduk, color = 'blue')
```

```python
#dari ke-3  soal terseut,akan saya sajikan contoh grafiknya, yaitu:
import matplotlib.pyplot as plt
%matplotlib inline

nama_desa_di_wilayah_karangploso = ['kepuharjo','ngijo','kendalsari','supiturang','kasin',]
jumlah_penduduk = ['87','88','95','77','99']

plt.figure(figsize=(7,3))
plt.bar(nama_desa_di_wilayah_karangploso, jumlah_penduduk, color = 'blue')

plt.title('jumlah warga yang terkena covid di wilayah karangploso', size=16)
plt.ylabel('jumlah warga yang terkena covid', size=14)
plt.xticks(size=12)
plt.yticks(size=12)

plt.show()
```

```python
#figure, axes, subplot
plt.figure()
ax1 = plt.axes()
ax1.plot(x,y)
ax1.set_title('Grafik sin(x)')

#perintah ini akan menampilkan grafik pada figure yang beda

plt.figure()
plt.plot(x,y)
plt.title('Grafik 2sin(x)')

plt.show()
```

```python
fig = plt.figure()
ax1 = plt.subplot(121)
ax1.plot(x,y)
ax1.set_title('Grafik sin(x)')
ax1.axis([0,6,-2,2])
ax1.set_xlabel('radian')
ax1.set_ylabel('sin(x)')

ax2 = plt.subplot(122)
ax2.plot(x,2*y,'r')
ax2.set_title('Grafik 2sin(x)')
ax2.axis([0,6,-2,2])
ax2.set_xlabel('radian')
ax2.set_ylabel('2sin(x)')
plt.show()
```
